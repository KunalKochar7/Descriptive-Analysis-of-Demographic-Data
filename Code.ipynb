### Task 1: Describe the frequency distributions of the variables. Consider also the differences between the sexes.

```python
sns.set(font_scale=1.2)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

dx = sns.histplot(df2022['total.fertility.rate'],stat='density', ax=axes[0, 0])
cx=sns.histplot(df2022['life.expectancy.both.sexes'],stat='density', ax=axes[0, 1])
ax=sns.histplot(df2022['life.expectancy.females'],stat='density', ax=axes[1, 1])
bx=sns.histplot(df2022['life.expectancy.males'],stat='density', ax=axes[1, 0])

cx.set(xlabel = 'Life Expectancy Both sexes', ylim=(0.00, 0.08),xlim=(50,100))
ax.set(xlabel = 'Life Expectancy Females' , ylim=(0.00, 0.08),xlim=(50,100))
bx.set(xlabel = 'Life Expectancy Males', ylim=(0.00, 0.08),xlim=(50,100))
dx.set(xlabel = 'Total Fertility Rate')
plt.savefig('Figure 1.pdf')
```

#### Differences between sexes

```python
a=plt.scatter(df2022['life.expectancy.both.sexes'], df2022['life.expectancy.females'], color='r')
b=plt.scatter(df2022['life.expectancy.both.sexes'], df2022['life.expectancy.males'], color='b')
plt.xlabel('Life Expectancy Range',size=12)
plt.ylabel('Life Expectancy Range',size=12)
plt.title('Differences between the sexes')
plt.legend((a,b),('Females','Males'))
plt.show()
```

### Task 2: Describe the dependence structure of the involved variables. Are there hints towards monotonic relationships between variables? Are there hints towards functional relationships that are monotone but not linear?


```python
df2022_new = df2022.drop(columns = ['country', 'subregion', 'region', 'year'])
```

```python
corr=df2022_new.corr().round(3)
corr
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.heatmap(corr,vmin=-1,vmax=1,annot=True,cmap="YlGnBu")
```

### Task 3: Describe the variablesâ€™ variabilities within and between subregions.


```python
sns.set(rc={'figure.figsize':(8,8)})
my_pal = sns.color_palette("bright")
ax = sns.boxplot(data=(df2022.sort_values(["region"])),x='total.fertility.rate', y='subregion', hue='region', palette=my_pal)
ax.set(xlabel='Total Fertility Rate', ylabel='Subregion')
```

```python
ax = sns.boxplot(data=(df2022.sort_values(["region"])),x='life.expectancy.both.sexes', y='subregion', hue='region', palette=my_pal)
ax.set(xlabel='Life Expectancy at Birth for Both Sexes', ylabel='Subregion')
```

```python
ax = sns.boxplot(data=(df2022.sort_values(["region"])),x='life.expectancy.males', y='subregion', hue='region', palette=my_pal)
ax.set(xlabel='Life Expectancy Males', ylabel='Subregion')
```

```python
ax = sns.boxplot(data=(df2022.sort_values(["region"])),x='life.expectancy.males', y='subregion', hue='region', palette=my_pal)
ax.set(xlabel='Life Expectancy Females', ylabel='Subregion')
```
